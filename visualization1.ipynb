```python
#Real data automobile_data.csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#to load dataset
df=pd.read_csv("automobile_data.csv")
```

```python
df.head()
```

```python
X=df["highway-mpg"]  #independent variable
Y=df["price"]   #dependent means target variable
#Scatter graph
plt.figure(figsize=(7,7))
plt.scatter(X,Y)
plt.title("Automobile comapny",color="Red",fontsize=20,fontname="Arial")
plt.xlabel("Highway milege",color="Green",fontsize=18)
plt.ylabel("price",color="Green",fontsize=18)
plt.show()
```

```python
df_num=df.select_dtypes(['int64','float64'])
df_num.columns   #To display column name of dataset df_num
```

```python
#here X is list which hold column name of dataset
X=['symboling','width', 'height', 'engine-size', 'city-mpg', 'highway-mpg']
Y=df_num["price"]
for i,col in enumerate(X):
#enumerate() inbuilt function : to retrieve index value of column at those index
    plt.figure(figsize=(6,6))
    plt.subplot(3,2,i+1)
    plt.scatter(df_num[col],Y)
    plt.xlabel(col)
    plt.ylabel("price")
```

```python
#Bar graph :
plt.figure(figsize=(7,8))
f=df["make"].value_counts()
df["make"].value_counts().plot(kind="bar")
plt.yticks(f)
plt.title("Automobile company")
plt.xlabel("Manufacuring company")
plt.ylabel("No of cars")
plt.show()
```

```python
#Bar graph :
plt.figure(figsize=(10,10))
f=df["body-style"].value_counts()
df["body-style"].value_counts().plot(kind="bar")
plt.yticks(f,rotation=90)
plt.title("Automobile company")
plt.xlabel("Body style of car")
plt.ylabel("No of cars")
plt.show()
```

```python
#pie graph
plt.figure(figsize=(7,7))
df["body-style"].value_counts().plot(kind="pie",autopct="%.1f%%")
#sedan formula : no of sedan body-style car/no of rows *100
#decimal : %f
plt.title("Automobile company")
plt.show()
```

```python
#pie graph
plt.figure(figsize=(7,7))
df["body-style"].value_counts().plot(kind="pie",autopct="%.1f%%",explode=(0,0,0.4,0,0.8))
#sedan formula : no of sedan body-style car/no of rows *100
#decimal : %f
plt.title("Automobile company")
plt.show()
```
